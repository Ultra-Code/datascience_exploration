```python
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv("./Dataset/rand-dollar.csv")
df.head()
```

```python
Y = df[["ZAR/USD"]]
X = df.drop("ZAR/USD",axis=1)
display(Y)
display(X)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train[["Value of Exports (ZAR)"]], Y_train)
```

```python
corr = df.corr(numeric_only=True)[["ZAR/USD"]]
corr = corr.drop("ZAR/USD") # avoid comparing with self
corr.sort_values(by="ZAR/USD", ascending=False)
```

```python
# Find the variable with the weakest linear relationship (smallest absolute correlation value)
weakest_variable = corr.abs().idxmin()

print(f"The variable with the weakest linear relationship with the 'ZAR/USD exchange rate' is: {weakest_variable}")
```

```python
strongest_variable = corr.idxmax()
print(f"{strongest_variable.tolist()[0]} has the strongest linear relationship with 'ZAR/USD'")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, Lasso
```

```python
X_scaled = StandardScaler().fit_transform(X.drop("Unnamed: 0",axis=1))
X_train_s, X_test_s, Y_train, Y_test = train_test_split(X_scaled,Y,test_size=0.2,shuffle=False)
```

```python
ridge = Ridge()
lasso = Lasso(alpha=0.01)
```

```python
ridge.fit(X_train_s,Y_train)
lasso.fit(X_train_s,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_train,ridge.predict(X_train_s))
```

```python
mean_squared_error(Y_train,lasso.predict(X_train_s))
```

```python
mean_squared_error(Y_test,ridge.predict(X_test_s))
```

```python
mean_squared_error(Y_test,lasso.predict(X_test_s))
```
