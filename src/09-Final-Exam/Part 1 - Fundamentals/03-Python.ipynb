```python
algeria_players = df[df['Nationality']=='Algeria']
algeria_players.sort_values(by=['Overall'], ascending=False).head()
```

```python
back_players = df[df['Preferred Positions Type']=='Back']
back_players.sort_values(by='Sliding tackle', ascending=False).head()
```

```python
england_palyers = df[df['Nationality']=='England']
england_palyers.groupby(by="Preferred Positions Type")[['Overall']].mean().sort_values(by='Overall', ascending=False)
```

```python
england_palyers = df[df['Nationality']=='Brazil']
england_palyers.groupby(by="Preferred Positions Type")[['Overall']].mean().sort_values(by='Overall', ascending=False)
```
