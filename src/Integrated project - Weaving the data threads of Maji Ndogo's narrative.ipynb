```python
%%sql -- Find discrepancies in data compared to what the auditor gave
-- 102 records were inaccurate while 1518 records were accurate meaning our data had 94% accuracy
SELECT a.location_id,
        v.record_id,
        a.true_water_source_score as auditor_score,
        q.subjective_quality_score as surveyor_score
FROM
    auditor_report as a
JOIN
    visits as v
ON
    a.location_id=v.location_id
JOIN
    water_quality as q
ON
    q.record_id=v.record_id
WHERE 
    a.true_water_source_score != q.subjective_quality_score
AND
    v.visit_count=1
ORDER BY location_id
```

```python
%%sql -- Meaning the water_quality data is authentic
SELECT a.location_id,
        v.record_id,
        a.type_of_water_source as auditor_source,
        w.type_of_water_source as surveyor_source,
        a.true_water_source_score as auditor_score,
        q.subjective_quality_score as surveyor_score
FROM
    auditor_report as a
JOIN
    visits as v
ON
    a.location_id=v.location_id
JOIN
    water_quality as q
ON
    q.record_id=v.record_id
JOIN
    water_source as w
ON 
    w.source_id=v.source_id
WHERE 
    a.true_water_source_score != q.subjective_quality_score
AND
    v.visit_count=1
AND a.type_of_water_source != w.type_of_water_source
ORDER BY location_id
```

```python
%%sql -- Find the employees with the most mistakes
WITH Incorrect_Records AS (
SELECT a.location_id,
        v.record_id,
        e.employee_name,
        a.true_water_source_score as auditor_score,
        q.subjective_quality_score as surveyor_score
FROM
    auditor_report as a
JOIN
    visits as v
ON
    a.location_id=v.location_id
JOIN
    water_quality as q
ON
    q.record_id=v.record_id
JOIN
    employee as e
ON
    v.assigned_employee_id = e.assigned_employee_id
WHERE 
    a.true_water_source_score != q.subjective_quality_score
AND
    v.visit_count=1
),
Error_Count AS (
SELECT 
    employee_name, COUNT(employee_name) as mistake_count
FROM
    Incorrect_Records
GROUP BY
    employee_name
ORDER BY
    mistake_count DESC
)

SELECT *
FROM 
    Error_Count
WHERE mistake_count > (
SELECT 
    AVG(mistake_count) as average_mistake_count
FROM
    Error_Count);
```

```python
%%sql -- Improve and convert some CTE's to Views
CREATE VIEW IF NOT EXISTS Incorrect_Records AS(
SELECT a.location_id,
        v.record_id,
        e.employee_name,
        a.true_water_source_score as auditor_score,
        q.subjective_quality_score as surveyor_score
FROM
    auditor_report as a
JOIN
    visits as v
ON
    a.location_id=v.location_id
JOIN
    water_quality as q
ON
    q.record_id=v.record_id
JOIN
    employee as e
ON
    v.assigned_employee_id = e.assigned_employee_id
WHERE 
    a.true_water_source_score != q.subjective_quality_score
AND
    v.visit_count=1
);

-- This CTE calculates the number of mistakes each employee made
WITH Error_Count AS ( 
SELECT 
    employee_name, COUNT(employee_name) as mistake_count
FROM
    Incorrect_Records /*
                Incorrect_records is a view that joins the audit report to the database
                for records where the auditor and
                employees scores are different*/
GROUP BY
    employee_name
ORDER BY
    mistake_count DESC
),
Suspected_List AS (
SELECT employee_name, mistake_count
FROM 
    Error_Count
WHERE
    mistake_count > (
SELECT 
    AVG(mistake_count) as average_mistake_count
FROM
    Error_Count)
)

SELECT * FROM
    Incorrect_Records
WHERE 
    employee_name in (SELECT employee_name FROM Suspected_List)
ORDER BY employee_name;
```

```python
%%sql -- Add auditor comments to table for investigating suspects
CREATE VIEW IF NOT EXISTS Incorrect_Records AS(
SELECT a.location_id,
        v.record_id,
        e.employee_name,
        a.true_water_source_score as auditor_score,
        q.subjective_quality_score as surveyor_score
FROM
    auditor_report as a
JOIN
    visits as v
ON
    a.location_id=v.location_id
JOIN
    water_quality as q
ON
    q.record_id=v.record_id
JOIN
    employee as e
ON
    v.assigned_employee_id = e.assigned_employee_id
WHERE 
    a.true_water_source_score != q.subjective_quality_score
AND
    v.visit_count=1
);

-- This CTE calculates the number of mistakes each employee made
WITH Error_Count AS ( 
SELECT 
    employee_name, COUNT(employee_name) as mistake_count
FROM
    Incorrect_Records /*
                Incorrect_records is a view that joins the audit report to the database
                for records where the auditor and
                employees scores are different*/
GROUP BY
    employee_name
ORDER BY
    mistake_count DESC
),
Suspected_List AS (
SELECT employee_name, mistake_count
FROM 
    Error_Count
WHERE
    mistake_count > (
SELECT 
    AVG(mistake_count) as average_mistake_count
FROM
    Error_Count)
)

-- This query filters all of the records where the "corrupt" employees gathered data.
SELECT 
    i.employee_name, i.location_id, a.statements
FROM
    Incorrect_Records as i
JOIN
    auditor_report as a
ON
    a.location_id=i.location_id
WHERE 
    employee_name in (SELECT employee_name FROM Suspected_List)
ORDER BY employee_name;
```

```python
%%sql -- Final evidence that these officials are corrupt
CREATE VIEW IF NOT EXISTS Incorrect_Records AS(
SELECT a.location_id,
        v.record_id,
        e.employee_name,
        a.true_water_source_score as auditor_score,
        q.subjective_quality_score as surveyor_score
FROM
    auditor_report as a
JOIN
    visits as v
ON
    a.location_id=v.location_id
JOIN
    water_quality as q
ON
    q.record_id=v.record_id
JOIN
    employee as e
ON
    v.assigned_employee_id = e.assigned_employee_id
WHERE 
    a.true_water_source_score != q.subjective_quality_score
AND
    v.visit_count=1
);

-- This CTE calculates the number of mistakes each employee made
WITH Error_Count AS ( 
SELECT 
    employee_name, COUNT(employee_name) as mistake_count
FROM
    Incorrect_Records /*
                Incorrect_records is a view that joins the audit report to the database
                for records where the auditor and
                employees scores are different*/
GROUP BY
    employee_name
ORDER BY
    mistake_count DESC
),
Suspected_List AS (
SELECT employee_name, mistake_count
FROM 
    Error_Count
WHERE
    mistake_count > (
SELECT 
    AVG(mistake_count) as average_mistake_count
FROM
    Error_Count)
)

-- This query filters all of the records where the "corrupt" employees gathered data.
SELECT 
    i.employee_name, i.location_id, a.statements
FROM
    Incorrect_Records as i
JOIN
    auditor_report as a
ON
    a.location_id=i.location_id
WHERE 
    employee_name in (SELECT employee_name FROM Suspected_List)
AND
    a.statements LIKE "%cash%"
ORDER BY employee_name;
```
